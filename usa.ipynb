```python
import pandas as pd
import numpy as nu
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data from file
usa_data = pd.read_csv("F:\\python\\data\\001\\data.csv")
```

```python
# show data set
usa_data
```

```python
# show data subsets type
usa_data.info()
```

```python
# show statistic summary
usa_data.describe()
```

```python
# verify if there is not-a-number in data subset
usa_data.isna()
```

```python
# add column with "squarness" parameter
usa_data['squarness'] = usa_data['length'] / usa_data['width']
```

```python
usa_data
```

```python
# sort values by household income
usa_data.sort_values(by="med_household_income")
```

```python
# make a copy of data set, as independent instance
usa_data_no_missing = usa_data.copy(deep=True)
```

```python
# remove not-a-number
usa_data_no_missing = usa_data_no_missing.dropna()
```

```python
usa_data_no_missing
```

```python
# remove all data which are not a numeric values
usa_data_no_missing = usa_data_no_missing[usa_data_no_missing.med_household_income.apply(lambda x: x.isnumeric())]
```

```python
usa_data_no_missing.info()
```

```python
usa_data_no_missing
```

```python
# change data type from object to int
usa_data_no_missing["med_household_income"] = usa_data_no_missing['med_household_income'].astype(int)
```

```python
# create new column based on calualtion
usa_data_no_missing["income_per_capita"] = usa_data_no_missing['med_household_income'] / usa_data_no_missing['total_pop']
```

```python
# remove column
usa_data_no_missing.drop(columns=['income_per_capita'])
```

```python
#plot scatter graph
plt.scatter(usa_data_no_missing['med_household_income'],usa_data_no_missing['squarness'])
```

```python
plt.scatter(usa_data_no_missing['med_household_income'],usa_data_no_missing['density'])
```

```python
plt.scatter(usa_data_no_missing['med_household_income'],usa_data_no_missing['area'])
```

```python
# calulate correlation
correlation = usa_data_no_missing.corr()
```

```python
correlation
```

```python
# present correlation as heatmap
map = sns.heatmap(correlation, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(50, 500, n=500),
square=True
)
plt.show()
```
